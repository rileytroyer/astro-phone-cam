## First define the image directories

```python
img_dir = '../images/'

light_dir = img_dir + 'lights/'
dark_dir = img_dir + 'darks/'
bias_dir = img_dir + 'biases/'
flat_dir = img_dir + 'flats/'
```

```python
# Import libraries

import numpy as np
import os
from skimage import io, exposure
from skimage.restoration import denoise_nl_means, estimate_sigma
from matplotlib import pyplot as plt
```

## Step 1: Average the images
### The first thing we need to do is to read in the images and average them
This is technically the image stacking step. It is actually very simple and not nearly as technical as it sounds. The issue with this basic approach is that if we stack too many images we will start to see star trails. More advanced programs account for this and shift each image slightly to counteract the planet's rotation.

Because phone images are very noisy I have taken the red, green, and blue color channels and averaged them all together to produce a grayscale image. For me this seemed to produce a nicer image.

```python
def average_images(image_directory, image_names):
    """Function to average a number of images stored in the 
    specified directory
    INPUT
    image_directory
        type: string
        about: where the images are stored
    image_names
        type: array of strings
        about: list of all the image names
    OUTPUT
    averaged_array
        type: array of floats
        about: the averaged image
    """
    # Create an empty array to store averaged array in
    tmp_image = io.imread(image_directory + image_names[0])
    size = tmp_image.shape
    averaged_array = np.zeros(size)

    # Loop through each file
    for n, image_name in enumerate(image_names):
        
        file = image_directory + image_name
        image_array = io.imread(file)
        
        # Add array values to average array
        if n == 0:
            averaged_array = np.zeros(image_array.shape)

        else:
            averaged_array = averaged_array + image_array

    # Divide summed array to get average
    averaged_array = averaged_array/n

    return averaged_array
```

```python
# Get list of light, dark, biases, and flat images
light_imgs = sorted(os.listdir(light_dir))
dark_imgs = os.listdir(dark_dir)
bias_imgs = os.listdir(bias_dir)
flat_imgs = os.listdir(flat_dir)
```

```python
# Get average image for lights
# 30 sec seems to be about the max before star trails
avg_light = average_images(light_dir, light_imgs[0:30])

# Average all of the color channels...MORE DATA!
avg_light = np.mean(avg_light, axis=2)

# Get average image for darks
avg_dark = average_images(dark_dir, dark_imgs)
avg_dark = np.mean(avg_dark, axis=2)

# Get average image for biases
avg_bias = average_images(bias_dir, bias_imgs)
avg_bias = np.mean(avg_bias, axis=2)

# Get average image for flats
avg_flat = average_images(flat_dir, flat_imgs)
avg_flat = np.mean(avg_flat, axis=2)
```

## Step 2: Calibration
### After stacking the images we can now calibrate them using the dark, bias, and flat frames
This isn't necessary and I'm still on the fence if this makes much of a difference for these phone images, but if you can I think it's still worth doing.

### Here is the formula for this calibration
$ \text{processed image} = \frac{\text{image} - \text{dark}}{\text{flat} - \text{bias}}$

```python
# Perform the calibration
calibrated_img = (avg_light - avg_dark) / (avg_flat - avg_bias)
```

## Step 3a: Denoise Processing
### We have our calibrated image, now we can start to process it
It is probably worth reducing the noise in the image. Even after the stacking, phone images are going to have a lot of noise. From my experience this is better to do before increasing the contrast.

The weight variable controls how strong the denoising is.

```python
# How much to denoise, feel free to play around with this
weight = 4

# Estimate the sigma of the image for noise
sigma_est = np.mean(estimate_sigma(calibrated_img))

# Keywords to specify the patch size for the denoise algorithm
patch_kw = dict(patch_size=5, 
                patch_distance=6)

processed_img = denoise_nl_means(calibrated_img, h=weight*sigma_est, 
                                 sigma=sigma_est,
                                 fast_mode=True,
                                 **patch_kw)
```

## Step 3b: Increase Contrast Processing
After reducing the noise we can now boost the contrast of the image to pull out the faint details.

The low and high variables control where the black point and white point will be. They are given in percentage of pixels.

```python
# The lowest percentage of the pixels that will become black
low = 5

# The highest percentage of pixels that will become white
high = 99.97

v_min, v_max = np.percentile(processed_img, (low, high))
processed_img = exposure.rescale_intensity(processed_img, 
                                           in_range=(v_min, v_max))
```

## Step 4: Display the image

```python
# Plot the image

# I needed to rotate mine
plot_img = np.rot90(processed_img, 2)

plt.figure(figsize=(16, 9))
plt.imshow(plot_img, cmap='gray')

plt.axis('off')
plt.tight_layout()

# Save image 
#plt.savefig('../processed_image.jpg', dpi=500)
```

## For reference here is what a single image looks like

```python
# Read in a single image
single_img = io.imread(light_dir + light_imgs[0])

# Average all color channels
single_img = np.mean(single_img, axis=2)

# Denoise
weight = 4
sigma_est = np.mean(estimate_sigma(single_img))
patch_kw = dict(patch_size=5, 
                patch_distance=6)
single_img = denoise_nl_means(single_img, h=weight*sigma_est, 
                                 sigma=sigma_est,
                                 fast_mode=True,
                                 **patch_kw)

# Increase contrast
low = 5
high = 99.97
v_min, v_max = np.percentile(single_img, (low, high))
single_img = exposure.rescale_intensity(single_img, 
                                           in_range=(v_min, v_max))

# Plot the image
single_img = np.rot90(single_img, 2)

plt.figure(figsize=(16, 9))
plt.imshow(single_img, cmap='gray')

plt.axis('off')
plt.tight_layout()
```
